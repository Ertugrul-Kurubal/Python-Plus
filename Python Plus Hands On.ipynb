```python
def calc_1(a, b, c):
  if c == "+":
    return(a+b)
  elif c == "-":
    return(a-b)
  elif c == "*":
    return(a*b)
  elif c == "/":
    return(a/b)
  else:
    print("Enter valid argument!")
    
print(calc_1(5, 2, "*")) 
```

```python
fruits = ["banana", "mango", "pear", "apple", "kiwi", "grape"]
count1 = 0
count2 = 3
while count1<3:
  try :
    x = int(input("Enter a index number "))
    if fruits[x] == "kiwi":
      print("My favourite fruit is {}".format(fruits[x]))
      break
  except:
    count2 -= 1
    print("You have {} right left. Try again".format(count2))
  else:
    print("Congrats! You've entered a valid input")
  finally:
    print("Our fruit are always fresh!")
  count1 += 1  
```
